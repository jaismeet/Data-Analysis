# Analysis of Udemy Courses

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

file = "udemy_courses.csv"
data = pd.read_csv("udemy_courses.csv")

# display the first few rows of the dataset
data.head()
```

```python
# display the last few rows of dataset
data.tail()
```

```python
# shape of the dataframe
print(f'Rows: {data.shape[0]}, Columns: {data.shape[1]}')
```

```python
# display the basic information  about the dataframe
print(data.info())
```

```python
#  Basic statistics about dataframe
data.describe(include = 'all')
```

```python
# checking for null values
data.isnull()
```

```python
data.isna().any()
```

```python
# check for duplicate data 
data.duplicated()
```

```python
# check for any duplicate data 
data.duplicated().any()
```

```python
# sum of duplicate values
data.duplicated().sum()
```

```python
data.duplicated().value_counts()
```

```python
# remove the duplicate rows
data.drop_duplicates(inplace = True)
data
```

```python
# shape of the dataframe after removing the duplicated rows
print(f'Rows: {data.shape[0]}, Columns: {data.shape[1]}')
```

```python
# change the values of is_paid column
data['is_paid'] = data['is_paid'].replace({True:"paid" , False:"unpaid"})
data.to_csv("udemy_courses.csv" , index = False)
data.head()
```

```python
# number of courses per subject 
num = data['subject'].value_counts()
num
```

```python
num.plot(kind = 'bar' , color = ['lightsalmon','lightgreen' , 'lightblue' ,'pink'])
plt.title("Number of Courses Per Subject")
plt.xlabel('Subjects')
plt.ylabel('Number of courses per subject')
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```

```python
# for which levels , udemy courses provides the courses
num = data[['level']].value_counts()
num
```

```python
sns.countplot(x = 'level' , data = data , palette='husl',hue ='level')
plt.title("Number of Courses Level Wise")
plt.xlabel('levels')
plt.ylabel('Number of courses per levels')
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```

```python
# display the count of paid courses and free courses
paid_count = data['is_paid'].value_counts()
paid_count.plot(kind = "pie" , autopct = "%1.1f%%" ,colors = ["lightcoral","palegreen"])
plt.title("Paid v/s Unpaid Courses Available")
plt.tight_layout()
plt.show()
```

```python
# display which course has more lectures (free or paid)
data['num_lectures'].groupby(data['is_paid']).sum().sort_values(ascending = False)
```

```python
sns.barplot(x = 'num_lectures',y = 'is_paid',data = data , palette = 'pastel' , hue = 'is_paid')
plt.title("Which Course Has More Lectures (paid or unpaid)")
plt.show()
```

```python
# display which course has a higher numbers of subscriber (free or paid)
data['num_subscribers'].groupby(data["is_paid"]).sum().sort_values(ascending = False)
```

```python
sns.barplot(x='is_paid' , y = 'num_subscribers' , data = data , palette='coolwarm' , hue = 'is_paid')
plt.title("Which Course Has A Higher Number Of Subscriber (unpaid or paid)")
plt.show()
```

```python
# find most popular course title
data['num_subscribers'].groupby(data['course_title']).sum().sort_values(ascending=False).head(1)
```

```python
data[data['num_subscribers']==data['num_subscribers'].max()]['course_title']
```

```python
# 10 most popular course title as per the number of subscribers
top10 = data.sort_values(by='num_subscribers', ascending=False).head(10)
top10
```

```python
sns.barplot(x = "num_subscribers" , y = "course_title" , data = top10 , palette="viridis" , hue = 'num_subscribers')
plt.title("Top 10 Most Popular Course Title As Per The No. Of Subscriber")
plt.show()
```

```python
# course which have highest number of reviews
data[data['num_reviews']==data['num_reviews'].max()]['subject']
```

```python
sns.barplot(x = 'subject' , y = 'num_reviews', data = data , palette='pastel' , hue = 'subject')
plt.xticks(rotation =45 )
plt.title("No. Of Reviews As Per The Courses")
plt.show()
```

```python
# does price affect the number of reviews 
sns.scatterplot(data=data, x='price', y='num_reviews' )
plt.figure(figsize=(15,10))
plt.show()
```

```python
# in which year highest number of courses were posted
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])
data['year'] = data['published_timestamp'].dt.year
data.head(1)
```

```python
data['year'].value_counts()
```

```python
sns.countplot(x='year' , data = data , palette='husl' , hue = 'year')
plt.title("No. Of Courses Assign By Year")
plt.show()
```

```python
# display categary wise count of posted subjects(year wise)
num =data.groupby(data['year'])['subject'].value_counts()
num
```

```python
num.plot(kind = 'bar',color='lightsalmon')
plt.xticks(rotation = 90 , fontsize = 9)
plt.title("Categary Wise Count Of Posted Subjects(year wise)")
plt.show()
```
